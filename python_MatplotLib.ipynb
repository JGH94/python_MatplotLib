```python
import matplotlib.pyplot as plt 
import numpy as np
from mpl_toolkits.mplot3d import Axes3D # 3D를 위한 설치
 
# 한글 깨짐 오류 수정
plt.rcParams['font.family'] ='Malgun Gothic' 
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 기초 사용법 
print("1. 기초 사용법")
plt.plot([1, 2, 3, 4]) # plt 리스트 하나일 경우 y로 인식, X값은 자동으로 생성
plt.show() 

# 라인 및 점선 표현
# plt 리스트가 2개는 첫번째 인자는 X, 두번째 인자가 Y 값을 표현
print("2. 라인 및 점선 표현")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-') # x좌표에  1, 4, 9 , 16을 표현 , b- 는 blue(파랑)의 -라인을 의미
plt.axis([0, 6, 0, 20])
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # x좌표에  1, 4, 9 , 16을 표현 , ro는 빨간색 (‘red’)의 원형 (‘o’) 마커를 의미
plt.axis([0, 6, 0, 20])
plt.show()

# 여러 개의 표현
print("3. 여러개의 표현")
num_ = np.arange(0., 5., 0.2)
plt.plot(num_, num_, 'r--', num_, num_**2, 'bs', num_, num_**3, 'g^')
plt.show()
```

```python
# 레이블의 표현
print("1. 레이블 표현")
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X_Lable', labelpad=15)
plt.ylabel('Y_Lable', labelpad=15)
plt.show()

# 레이블 폰트 설정
print("2. 레이블 폰트 설정 및 범례표시")
# 범례를 위한 lable
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Count')
font1 = {'family': 'serif',
         'color': 'red',
         'weight': 'bold',
         'size': 14
         }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
         }
plt.xlabel('X_Lable', labelpad=15, fontdict = font1)
plt.ylabel('Y_Lable', labelpad=15, fontdict = font2)
# 범례 표시 loc = 위치 설정
plt.legend(loc ='lower right')
plt.show() 
```

```python
# 여러 곡선 그리기
x = np.arange(0,2,0.2)
print(x)
print("1. 곡선 및 그리드 표기방법")
plt.plot(x,x,'b--',x,x**2,'ro',x,x**3,'g-') 
#그리드 표시
plt.grid(True)  
plt.show()

# Y 그리드만 표시 -- y를 x로 바꾸면 x좌표만 표기됨
x = np.arange(0,2,0.2)   
print("2. y 그리드만 표기")
plt.plot(x,x,'b--',x,x**2,'ro',x,x**3,'g-') 
#그리드 표시
plt.grid(True, axis='y')
plt.show()


# 그리드 스타일 표기
x = np.arange(0,2,0.2)   
print("3. 그리드 스타일 표기 방법")
plt.plot(x,x,'b--',x,x**2,'ro',x,x**3,'g-') 
#그리드 표시
plt.grid(True, axis='y', color = 'blue', alpha = 0.5, linestyle = '--')
# Title 표기 왼쪽, 오른쪽 등 폰트 설정 가능
plt.title('TITLE', loc='right', pad=20) 
title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('TITLE_2', fontdict=title_font, loc='left', pad=10)
plt.show()
```

```python
# 막대 그래프 표현

x = np.arange(3) # 3개 생성 
years = ['2020', '2021', '2022'] # X축
values = [500, 400, 700] # Y축
color = ['r','g','b']
plt.bar(x, values, color = color, width = 0.5) # 값 설정 및 색상표시, 굵기 설정
plt.title('Vertical', fontdict=title_font, loc='left', pad=10)  
# plt.bar(x, values, color = color, width = 1, edgecolor='lightgray', linewidth=5) 바 테두리 색상 및 두께 설정
plt.xticks(x, years)

plt.show()


# 막대 수평 그래프 표현

x = np.arange(3) # 3개 생성 
years = ['2020', '2021', '2022'] # X축
values = [500, 400, 700] # Y축
color = ['r','g','b']
plt.barh(x, values, color= color, height = 0.5) # 값 설정 및 색상표시, 굵기 설정
plt.title('horizontal', fontdict=title_font, loc='left', pad=10)
# plt.barh(x, values, color= color, height = 0.5, edgecolor='lightgray', linewidth=5) 바 테두리 색상 및 두께 설정
plt.yticks(x, years)

plt.show()
```

```python
# 산점도 그리기  scatter() 이용
np.random.seed(0) # 0~1사이의 값으로 설정
ran_ = 30         # 30개의 랜덤값  생성
x = np.random.rand(ran_)
y = np.random.rand(ran_)

print(x)
print(y)

plt.scatter(x,y)
plt.title('1. Scatter', loc = 'left')
plt.show()

# 크기와 색상을 설정할 수 있음
area = (30 * np.random.rand(ran_))**2 # 크기 설정 값
colors = np.random.rand(ran_)         # 색상 설정 값 
plt.scatter(x,y, c = colors, s = area, alpha=0.5, cmap='Spectral') # c = 색상, s = 크기, alpha = 투명도, cmap = 컬러맵 문자열
plt.colorbar() # 오른쪽 color bar
plt.title('2. Scatter Size And Color ', loc = 'left')
plt.show()


# 3차원 산점도 
# x, y, z 의 값의 범위 이해 필요
n = 50
xmin, xmax, ymin, ymax, zmin, zmax = 0, 15, 0, 15, 0, 40 # x, y, z 크기 설정
cmin, cmax = 0, 3 # color 


xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6)) # 3D 크기 설정
ax = fig.add_subplot(111, projection='3d') # 3D 만들기위한 설정
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')
plt.title("3. 3D Plt")
plt.show()
```

```python

# 그래프 객체화 
x = np.arange(1, 5)     # [1, 2, 3, 4] 
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)

# 그래프를 인덱스로 접근해서 값 적용
# 스타일이나 그래프 관련 디자인 변경도 가능
ax[0][0].plot(x, np.sqrt(x))      # left-top
ax[0][1].plot(x, x)               # right-top
ax[1][0].plot(x, -x+5)            # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5))   # right-bottom
  
plt.show()
```

```python
# Gradient descent 표시
fig, ax = plt.subplots()

ax.set_title('Mean Squared Error', pad=20)
ax.set_xlim(-3, 3) # x 좌표 값 세팅
ax.set_ylim(0, 3)  # y 좌표 값 세팅
ax.set_xticks([-3, -2, -1, 0, 1, 2, 3]) # x좌표 라인 표시
ax.set_yticks([1, 2, 3]) # y좌표 라인 표시

ax.spines['left'].set_position('center')        # 왼쪽 축을 가운데 위치로 이동
ax.spines['right'].set_visible(False)          # 오른쪽 축을 보이지 않도록
ax.spines['top'].set_visible(False)            # 위 축을 보이지 않도록
ax.spines['bottom'].set_position(('data', 0))   # 아래 축을 데이터 0의 위치로 이동
ax.tick_params('both', length=0)                # Tick의 눈금 길이 0

x = np.linspace(-3, 3,100) # x 좌표까지의 값, y좌표 까지의 값, 곡선의 정밀도
ax.set_xlabel('y_true - y_pred', fontdict={'fontsize': 14}, labelpad=10)
ax.plot(x, x**2, color= 'red', linewidth=5)

plt.show()
```

```python
# 이중 y축 표시 및 범례표시
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1 #[1,2,3]
y2 = -x - 1#[-1,-2,-3] 

fig, ax1 = plt.subplots() 
ax1.plot(x, y1, color='green') 
ax1.set_xlabel('X-Value')
ax1.set_ylabel('y1-Value')
line1 = ax1.plot(x, y1, color='green', label='y1 Data') # 범례 표시

ax2 = ax1.twinx()
ax2.set_ylabel('y2-Value') 
ax2.plot(x, y2, color='deeppink')
line2 = ax2.plot(x, y2, color='deeppink', label='y2 Data') # 범례 표시

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='lower right')
plt.show()
```

```python
# 두종류의 그래프 표시 set_zorder(), zorder
# 3단계로 정리 ( 기본 설정, 데이터 설정, 그래프 표시)

#1. 기본 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

#2. 데이터 설정 
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

#3. 그래프 표시
fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim(0, 18)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price ($)')
ax1.tick_params(axis='both', direction='in')
 
ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim(0, 18)
ax2.set_ylabel(r'Demand ($\times10^6$)')
ax2.tick_params(axis='y', direction='in')
 

# 범례 표시
ax1.set_zorder(ax2.get_zorder() + 10)
ax1.patch.set_visible(False) 
ax1.legend(loc='upper left')
ax2.legend(loc='upper right') 

plt.show()
```
